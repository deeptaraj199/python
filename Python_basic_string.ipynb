```python
#Create a program that takes a sentence from the user and counts the number of vowels (a, e, i, o, u) in the string
sentence = input("Enter a sentence:")
count = 0
vowels = set("aeiou")
sentence = sentence.lower()

for char in sentence:
    if char in vowels:
        count += 1

print("Number of vowels in the sentence:",count)
```

```python
#Given a string, reverse the order of characters using string slicing and print the reverse
string = input("Enter the string:")
reversed_string = string[::-1]
print("The reversed string is:", reversed_string)
```

```python
#Write a program that takes a string as input and checks if it is a palindrome
string = input("Enter the string:")
cleaned_string = string.replace(" ", " ").lower()
reversed_string = cleaned_string[::-1]

if cleaned_string == reversed_string:
  print("The string is a palindrome")
else:
  print("The string is not a palindrome")
```

```python
#Create a program that takes a string as input and removes all the spaces from it
string = input("Enter the string:")
no_spaces_string = "".join(string.split())
print("String with spaces removed:", no_spaces_string)
```
